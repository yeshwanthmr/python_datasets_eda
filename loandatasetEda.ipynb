```python
# fill the null values of numerical data type
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mean())
```

```python
df['LoanAmount'].isnull().sum()
df['Loan_Amount_Term'].isnull().sum()
df['Credit_History'].isnull().sum()
```

```python
#fill the null values in categorical datatype
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df.isnull().sum()
```

```python
print("Number of People who took loan by gender")
print(df['Gender'].value_counts())
sns.countplot(x='Gender',data=df,color="green")
plt.show()
```

```python
print("Number of People who took loan by married")
print(df['Married'].value_counts())
sns.countplot(x='Married',data=df,color="green")
plt.show()
```

```python
print("Number of People who took loan by Education")
print(df['Education'].value_counts())
sns.countplot(x='Education',data=df,color="green")
plt.show()
```

```python
# total applicant income
df['Total_Applicant'] = df['ApplicantIncome']+df['CoapplicantIncome']
df.head()
```

```python
# Applt log transformation for ApplicantIncome
df['ApplicantIncomelog'] = np.log(df['ApplicantIncome']+1)
sns.displot(df['ApplicantIncomelog'])
plt.show()
```

```python
# Applt log transformation for Loan Amount
df['LoanAmountlog'] = np.log(df['LoanAmount']+1)
sns.displot(df['LoanAmountlog'])
plt.show()
```

```python
# Applt log transformation for Loan Amount Term
df['Loan_Amount_Term_log'] = np.log(df['Loan_Amount_Term']+1)
sns.displot(df['Loan_Amount_Term_log'])
plt.show()
```

```python
# Applt log transformation for TotalIncome
df['TotalIncomelog'] = np.log(df['Total_Applicant']+1)
sns.displot(df['TotalIncomelog'])
plt.show()
```

```python
#drop unnecessary columns
cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Applicant','Loan_ID']
df.drop(cols,axis=1)
```

```python
# convert the categorical to numerical columns(labelencode,onehotencoder)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status']
for col in cols:
    df[col]=le.fit_transform(df[col])
```
